```python
import sqlite3
from typing import Optional, Dict

def buscar_questao_por_id(nome_do_banco: str, questao_id: int) -> Optional[Dict]:
    """
    Busca uma quest√£o no banco de dados pelo seu ID e retorna os dados como um dicion√°rio.
    """
    db_file = f'{nome_do_banco}.db'
    
    try:
        with sqlite3.connect(db_file) as conn:
            # Garante que as linhas sejam retornadas como objetos acess√≠veis por nome da coluna
            conn.row_factory = sqlite3.Row  
            cursor = conn.cursor()
            
            # Seleciona todos os campos
            cursor.execute("SELECT * FROM questoes WHERE id = ?", (questao_id,))
            
            registro = cursor.fetchone()
            
            if registro:
                # Converte o objeto Row para um dicion√°rio padr√£o
                return dict(registro) 
            else:
                return None
                
    except sqlite3.Error as e:
        print(f"‚ùå Erro ao buscar quest√£o no banco de dados: {e}")
        return None

NOME_BANCO = "BANCO_TESTE"  # Use o nome do seu banco
ID_DA_QUESTAO = 1                           # ID da quest√£o que voc√™ quer ver

# --- 1. Busca os dados no banco ---
dados_questao = buscar_questao_por_id(NOME_BANCO, ID_DA_QUESTAO)

if dados_questao:
    # --- 2. Exibe os metadados (para contexto) ---
    print(f"--- Quest√£o ID: {dados_questao['id']} ---")
    print(f"S√©rie: {dados_questao['serie']} | Dificuldade: {dados_questao['dificuldade']}")
    
    # --- 3. Acessa e Exibe o Texto da Quest√£o ---
    # A chave do dicion√°rio deve ser 'texto' (nome da coluna)
    texto_completo = dados_questao['texto']
    
    print("\n----------------------")
    print("  TEXTO DA QUEST√ÉO")
    print("----------------------")
    print(texto_completo)
    print("----------------------")
    
else:
    print(f"Quest√£o com ID {ID_DA_QUESTAO} n√£o encontrada no banco de dados.")
```

# Adicionar todas as quest√µes de um pdf de uma s√≥ vez

Adicionar todas as quest√µes de um pdf no banco de dados, utilizando um arquivo csv auxiliar onde todas as informa√ß√µes necess√°rias j√° est√£o dispon√≠veis.

Informa√ß√µes necess√°rias no csv:
 - n√∫mero da quest√£o
 - serie
 - origem
 - dificuldade
 - imagem
 - temas

Organizando o teste:

1. Ler o arquivo .csv
2. Extrair corretamente as informa√ß√µes do .csv
3. Utilizar um loop para adicionar o texto da quest√£o + infos do .csv no banco de dados.
4. Checar o banco de dados.

```python
import pandas as pd
import csv
from typing import List, Dict, Any, Optional
from criar_banco import insert_question, extract_questions_from_OBFEP, Questao

# Supondo que as classes e fun√ß√µes a seguir estejam definidas ou importadas:
# - class Questao: (com to_tuple, etc.)
# - def insert_question_poo(banco: str, questao: Questao):
# - def extract_questions_from_pdf_corrected(pdf_path: str, delimiter: str):

def popular_banco_com_classificacao(
    banco_nome: str,
    pdf_path: str,
    csv_path: str,
    pdf_delimiter: str = "Q."
) -> None:
    """
    L√™ quest√µes de um PDF e metadados de um CSV, mapeia os dados
    e insere as quest√µes classificadas no banco de dados.

    Parameters
    ----------
    banco_nome : str
        Nome do banco de dados (ex: 'meu_banco_questoes').
    pdf_path : str
        Caminho para o arquivo PDF contendo os textos das quest√µes.
    csv_path : str
        Caminho para o arquivo CSV contendo os metadados de classifica√ß√£o.
    pdf_delimiter : str
        Delimitador usado no PDF para iniciar uma nova quest√£o (ex: "Q.").
    
    Returns
    -------
    None
    """
    
    print("--- INICIANDO FLUXO DE MAPEAMENTO E INSER√á√ÉO (CSV + PDF) ---")
    
    # --- 1. Leitura e Prepara√ß√£o dos Metadados (CSV) ---
    try:
        # Carrega o CSV usando Pandas para f√°cil manipula√ß√£o
        df = pd.read_csv(csv_path)
        
        # O cabe√ßalho deve ser: n√∫mero_questao, serie, origem, dificuldade, imagem, tema1, tema2, tema3
        
        # Cria um dicion√°rio de metadados, usando o n√∫mero da quest√£o como chave
        # Isso permite o mapeamento r√°pido com o √≠ndice do PDF
        metadata_map: Dict[int, Dict[str, Any]] = {}
        
        for index, row in df.iterrows():
            # A chave de mapeamento deve ser o n√∫mero da quest√£o
            try:
                # O n√∫mero da quest√£o no CSV (que deve ser 1, 2, 3...)
                num_questao = int(row['numero_questao']) 
            except ValueError:
                print(f"‚ö†Ô∏è Aviso: 'numero_questao' inv√°lido na linha {index + 2} do CSV. Ignorando linha.")
                continue

            # Constr√≥i a lista de temas (ignorando c√©lulas vazias/NaN)
            temas = [
                str(t) for t in [row['tema1'], row['tema2'], row['tema3']] 
                if pd.notna(t) and str(t).strip() and str(t).lower() != 'none'
            ]
            
            # Armazena os metadados
            metadata_map[num_questao] = {
                'serie': row.get('serie'),
                'origem': row.get('origem'),
                'dificuldade': row.get('dificuldade'),
                'imagem_path': row.get('imagem'),
                'temas': temas
            }
        
        print(f"‚úÖ CSV Lido: {len(metadata_map)} metadados prontos para mapeamento.")

    except FileNotFoundError:
        print(f"‚ùå Erro: Arquivo CSV '{csv_path}' n√£o encontrado.")
        return
    except KeyError as e:
        print(f"‚ùå Erro de coluna: Verifique se o CSV tem o cabe√ßalho correto, faltando a coluna {e}.")
        return

    # --- 2. Leitura dos Textos das Quest√µes (PDF) ---
    textos_questoes: List[str] = extract_questions_from_OBFEP(pdf_path, pdf_delimiter)
    
    if not textos_questoes:
        print(f"‚ùå Erro: N√£o foi poss√≠vel extrair quest√µes do PDF '{pdf_path}'.")
        return
    
    print(f"‚úÖ PDF Lido: {len(textos_questoes)} textos de quest√µes extra√≠dos.")

    # --- 3. Mapeamento, Cria√ß√£o de Objeto e Inser√ß√£o no Banco ---
    
    total_inserido = 0
    
    # Itera√ß√£o sobre a lista de textos. O √≠ndice (i) + 1 √© o n√∫mero da quest√£o.
    for i, texto_questao in enumerate(textos_questoes):
        num_questao = i + 1 # Quest√£o 1, 2, 3...
        
        # Pula a inser√ß√£o se o metadado n√£o existir para esta quest√£o (ex: se o CSV for menor que o PDF)
        if num_questao not in metadata_map:
            print(f"‚ö†Ô∏è Aviso: Quest√£o {num_questao} do PDF n√£o tem metadados no CSV. Pulando.")
            continue
            
        metadata = metadata_map[num_questao]
        
        try:
            # Cria√ß√£o do objeto Questao com os dados completos
            nova_questao = Questao(
                texto=texto_questao.strip(), # Limpa o texto
                serie=metadata.get('serie', 'N/A'),
                origem=metadata.get('origem', 'PDF'),
                dificuldade=metadata.get('dificuldade', 'N/A'),
                imagem_path=metadata.get('imagem_path'),
                temas=metadata.get('temas')
            )
            
            # 4. Inser√ß√£o no banco de dados
            insert_question(banco_nome, nova_questao)
            total_inserido += 1
            
        except Exception as e:
            print(f"‚ùå Erro ao processar/inserir Quest√£o {num_questao}: {e}")
            
    print(f"\n--- FLUXO CONCLU√çDO. Total de {total_inserido} quest√µes inseridas/atualizadas. ---")
```
