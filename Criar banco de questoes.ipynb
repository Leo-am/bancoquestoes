## Visualizar Banco de Questões

```python
NOME_BANCO = "questoes_OBFEP"
```

```python
import sqlite3
import pandas as pd
conn = sqlite3.connect(f'{NOME_BANCO}.db')
query = "SELECT * FROM questoes"
df = pd.read_sql_query(query, conn)
# Exibe o DataFrame como uma tabela no Jupyter Lab
df.head()  # ou df para visualizar a tabela completa
```

```python
tema_especifico = "optica"
df_filtrado = df[df['tema'] == tema_especifico]
df_filtrado.head()
```

```python
def visualizar_questao(id_questao):
    # Conecta ao banco de dados SQLite
    conn = sqlite3.connect(f'{NOME_BANCO}.db')
    
    # Consulta SQL para buscar a questão pelo ID
    query = f"""
    SELECT serie, tema, dificuldade, texto 
    FROM questoes 
    WHERE id = {id_questao}
    """
    
    # Executa a consulta e carrega o resultado em um DataFrame
    df = pd.read_sql_query(query, conn)
    
    # Fecha a conexão com o banco de dados
    conn.close()
    
    # Verifica se a questão foi encontrada
    if df.empty:
        print(f"Nenhuma questão encontrada com o ID {id_questao}.")
        return
    
    # Extrai os dados da questão
    ano = df['serie'].iloc[0]
    tema = df['tema'].iloc[0]
    dificuldade = df['dificuldade'].iloc[0]
    texto = df['texto'].iloc[0]
    
    # Formata e exibe o resultado
    print(f"Ano: {ano}")
    print(f"Tema: {tema}")
    print(f"Dificuldade: {dificuldade}")
    print(f"Texto: {texto}")

# Exemplo de uso
visualizar_questao(1)
```
